## **CLUSTERING ALGORITHM**

```python
# Encode the gender column
label_encoder = LabelEncoder()
mall['gender_encoded'] = label_encoder.fit_transform(mall['Genre'])
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Assuming you have a DataFrame named "mall" containing your customer data
# with columns "gender_encoded", "Age", "Annual Income (k$)", and "Spending Score (1-100)"

# Select the features for clustering
features = mall[['gender_encoded', 'Age', 'Annual Income (k$)', 'Spending Score (1-100)']]

# Define the range of cluster numbers to try
min_clusters = 1
max_clusters = 10

# Initialize an empty list to store the WCSS values
wcss = []

# Iterate over the range of cluster numbers
for num_clusters in range(min_clusters, max_clusters+1):
    # Create an instance of KMeans algorithm
    kmeans = KMeans(n_clusters=num_clusters, random_state=42)
    
    # Fit the data to the KMeans algorithm
    kmeans.fit(features)
    
    # Get the within-cluster sum of squares (WCSS)
    wcss.append(kmeans.inertia_)

# Plot the WCSS values against the number of clusters
plt.plot(range(min_clusters, max_clusters+1), wcss, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.title('Elbow Method')
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import numpy as np

# Assuming you have a DataFrame named "mall" containing your customer data
# with columns "gender_encoded", "Age", "Annual Income (k$)", and "Spending Score (1-100)"

# Select the features for clustering
features = mall[['gender_encoded', 'Age', 'Annual Income (k$)', 'Spending Score (1-100)']]

# Define the range of cluster numbers to try
min_clusters = 2
max_clusters = 10

# Initialize an empty list to store the silhouette scores
silhouette_scores = []

# Iterate over the range of cluster numbers
for num_clusters in range(min_clusters, max_clusters+1):
    # Create an instance of KMeans algorithm
    kmeans = KMeans(n_clusters=num_clusters, random_state=42)
    
    # Fit the data to the KMeans algorithm
    kmeans.fit(features)
    
    # Get the cluster labels for each data point
    cluster_labels = kmeans.labels_
    
    # Calculate the silhouette score
    silhouette_avg = silhouette_score(features, cluster_labels)
    silhouette_scores.append(silhouette_avg)

# Find the index of the maximum silhouette score
optimal_num_clusters = np.argmax(silhouette_scores) + min_clusters

# Print the optimal number of clusters
print("Optimal number of clusters:", optimal_num_clusters)
```
